```python
import pandas as pd
import re


df = pd.read_csv('recipes.csv')

# 定义一个函数，将时间格式转换为分钟
def parse_cook_time(cook_time):
    if pd.isnull(cook_time):
        return 0  # 如果是空值，则返回0分钟

    hours_match = re.search(r'(\d+)H', cook_time)
    minutes_match = re.search(r'(\d+)M', cook_time)

    hours = int(hours_match.group(1)) if hours_match else 0
    minutes = int(minutes_match.group(1)) if minutes_match else 0

    total_minutes = hours * 60 + minutes
    return total_minutes

# 应用函数到CookTime列
df['CookTimeInMinutes'] = df['CookTime'].apply(parse_cook_time)
df['PrepTimeInMinutes'] = df['PrepTime'].apply(parse_cook_time)
df['TotalTimeInMinutes'] = df['TotalTime'].apply(parse_cook_time)
# 打印结果
df.head()
```

```python
columns_to_drop = ['CookTime', 'PrepTime', 'TotalTime']
df = df.drop(columns=columns_to_drop)
df.head()
```

```python
columns_to_drop = ['Description','Images','RecipeYield','RecipeInstructions']
df = df.drop(columns=columns_to_drop)
df.head()
```
