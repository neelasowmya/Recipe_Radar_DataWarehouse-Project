```python
import pandas as pd
df = pd.read_csv('bq-results-20231204-221510-1701728123414.csv')
print(df.describe())
df.head()
```

```python
columns_to_drop = ['RecipeId', 'Name', 'DatePublished']
df = df.drop(columns=columns_to_drop)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Build LabelEncoder isntance
label_encoder = LabelEncoder()

# Label Encoding，and repalcement
df['RecipeCategory'] = label_encoder.fit_transform(df['RecipeCategory'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

target_variable = 'AggregatedRating'

# Split the dataset into training and testing (80:20 split)
train_data, test_data = train_test_split(df, test_size=0.2, random_state=15)
df.head()
```

```python
# Extract features and target variable
X_train = train_data.drop(columns=[target_variable])
y_train = train_data[target_variable]

X_test = test_data.drop(columns=[target_variable])
y_test = test_data[target_variable]
# Build a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
# Make predictions on the test set
y_pred = model.predict(X_test)

# Compute Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
# 计算决定系数
r2 = r2_score(y_test, y_pred)


print(f"Mean Absolute Error: {mae}")
print(f"R² Score: {r2}")
```

```python
# Get model coefficients
coefficients = model.coef_

# Show coefficients for each feature
for feature, coef in zip(X_train.columns, coefficients):
    print(f"{feature}: {coef}")
```
